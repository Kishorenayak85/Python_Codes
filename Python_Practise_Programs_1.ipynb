```python
print('=====================STRING PROGRAMS==========Lost Some Due to restart before saving========')

##Find words which are greater than given length k
print('=========Find words which are greater than given length k')

#Input : str = "hello geeks for geeks 
#          is computer science portal" 
#        k = 4
#Output : hello geeks geeks computer 
#         science portal

input_str = "hello geeks for geeks is computer science portal"
k = 4

mod_str = input_str.split(' ')
c = 1
for i in mod_str:
    lnth = len(i)
    if lnth > k and c == 1:
        print('length is grteater than 4, and the word is : ', i)
        fnl_line = i
    elif lnth > k and c > 1:
        print('length is grteater than 4, and the word is : ', i)
        fnl_line = fnl_line + " " + i
    else:
        print('length is not grteater than 4, and the word is : ', i)
    c += 1

print(fnl_line)

##Python program for removing i-th character from a string
print('====Python program for removing i-th character from a string')

#g e e k s
#0 1 2 3 4
#Input : Geek
#        i = 1
#Output : Gek

ip_str = 'my name is covid and i am deadly'
remove = 5
fnl_str = str_1 + str_2
print('before removeing:\n',ip_str,'\n after removing:\n',fnl_str )

##Python program to split and join a string
print('=====Python program to split and join a string')
# function split() to split a string and join() 
#Input : Geeks for Geeks
#Output : ['Geeks', 'for', 'Geeks']

ip_str = 'Geeks for Geeks'
mod_str = ip_str.split(' ')
c = 1
for i in mod_str:
    lnth = len(i)
    if c == 1:
        #print('Loop IP : ', i)
        fnl_line = i
    elif c > 1:
        #print('Loop IP : ', i)
        fnl_line = fnl_line + " : " + i
    else:
        print('IGNORE', i)
    c += 1

print(fnl_line)


def split_string(string): 
  
    # Split the string based on space delimiter 
    list_string = string.split(' ') 
      
    return list_string 
  
def join_string(list_string): 
  
    # Join the string based on '-' delimiter 
    string = '-'.join(list_string) 
      
    return string 
  
# Driver Function 
if __name__ == '__main__': 
    string = 'Geeks for Geeks'
      
    # Splitting a string 
    list_string = split_string(string) 
    print(list_string) 
  
     # Join list of strings into one 
    new_string = join_string(list_string) 
    print(new_string) 
    
    
##Python | Check if a given string is binary string or not
print('=========Python | Check if a given string is binary string or not')
#Input: str = "01010101010"
#Output: Yes

ip_str = "adlkda010101"
mod_str = set(ip_str)
c=0
for i in mod_str:
    if i not in ('0','1'):
        c += 1
print(c)
if c > 0:
    print(ip_str,' is not binary')
else:
    print(ip_str,' is binary')
    
##Python | Find all close matches of input string from a list
print('============Python | Find all close matches of input string from a list')

#Input : patterns = ['ape', 'apple', 'peach', 'puppy'], 
#          input = 'appel'
#Output : ['apple', 'ape']
##difflib.get_close_matches(word, possibilities, n, cutoff)
from difflib import get_close_matches 
  
def closeMatches(patterns, word): 
     print(get_close_matches(word, patterns)) 
  
# Driver program 
if __name__ == "__main__": 
    word = 'appel'
    patterns = ['ape', 'apple', 'peach', 'puppy'] 
    closeMatches(patterns, word) 
    
##Python program to find uncommon words from two Strings
print('=======Python program to find uncommon words from two Strings')

#Input : A = "Geeks for Geeks" 
#        B = "Learning from Geeks for Geeks"
#Output : ['Learning', 'from']

#Input : A = "apple banana mango" 
#        B = "banana fruits mango"
#Output : ['apple', 'fruits']

A = "Geeks for Geeks" 
B = "Learning from Geeks for Geeks"
#A = "apple banana mango" 
#B = "banana fruits mango"
mod_A = A.split(' ')
mod_B = B.split(' ')
#print(mod_A)
#print(mod_B)
mer_A_B = mod_A + mod_B
print(mer_A_B)
fnl_lst = []
for i in mer_A_B:
    c = 0
    #print('current word: ',i)
    for k in mer_A_B:
        if i == k:
            c +=1
    if c > 1:
        fnl_lst.append(i)
print('Common are: ',fnl_lst)
unc_cmn = []
for i in mer_A_B:
    if i not in fnl_lst:
        unc_cmn.append(i)
print('Uncommon words are: ',unc_cmn)    

##Python | Swap commas and dots in a String
print('=====Python | Swap commas and dots in a String')

#Input : 14, 625, 498.002
#Output : 14.625.498, 002

ip_str = "14, 625, 498.002"
ip_str.replace(',','~')
ip_str.replace('.',',')
ip_str.replace('~','.')

##Python | Permutation of a given string using inbuilt function
print('====\n===PENDING==Python | Permutation of a given string using inbuilt function\n===PENDING===')

##Python | Check for URL in a String
print('======Python | Check for URL in a String====')

#Input : string = 'My Profile: 
#https://auth.geeksforgeeks.org/user/Chinmoy%20Lenka/articles 
#in the portal of http://www.geeksforgeeks.org/'

#Output : URLs :  ['https://auth.geeksforgeeks.org/user/Chinmoy%20Lenka/articles',
#'http://www.geeksforgeeks.org/']

#Input : string = 'I am a blogger at https://geeksforgeeks.org'
#Output : URL :  ['https://geeksforgeeks.org']

import re 
  
def Find(string): 
    # findall() has been used  
    # with valid conditions for urls in string 
    url = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+] |[!*\(\), ]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', string) 
    return url 
      
# Driver Code 
string = 'My Profile: https://auth.geeksforgeeks.org / user / Chinmoy % 20Lenka / articles in the portal of http://www.geeksforgeeks.org/' 
print("Urls: ", Find(string))

##Execute a String of Code in Python
print('======Execute a String of Code in Python')

#Input:
#code = """ a = 6+5
#           print(a)"""
#Output:
#11
#Explanation:
#Mind it that "code" is a variable and
#not python code. It contains another code, 
#which we need to execute.

# function illustrating how exec() functions. 
def exec_code(): 
    LOC = """ 
def factorial(num): 
    fact=1 
    for i in range(1,num+1): 
        fact = fact*i 
    return fact 
print(factorial(5)) 
"""
    exec(LOC) 
      
# Driver Code 
exec_code() 


##String slicing in Python to rotate a string
print('=====String slicing in Python to rotate a string')

#Given a string of size n, write functions to perform following operations on string.#

#Left (Or anticlockwise) rotate the given string by d elements (where d <= n).
#Right (Or clockwise) rotate the given string by d elements (where d <= n).

#Input : s = "GeeksforGeeks"
#        d = 2
#Output : Left Rotation  : "eksforGeeksGe" 
#         Right Rotation : "ksGeeksforGee"  

def rotate(input,d): 
  
    # slice string in two parts for left and right 
    Lfirst = input[0 : d] 
    Lsecond = input[d :] 
    Rfirst = input[0 : len(input)-d] 
    Rsecond = input[len(input)-d : ] 
  
    # now concatenate two parts together 
    print("Left Rotation : ", (Lsecond + Lfirst) )
    print("Right Rotation : ", (Rsecond + Rfirst) )
# Driver program 
if __name__ == "__main__": 
    input = 'GeeksforGeeks'
    d=2
    rotate(input,d) 
    
##String slicing in Python to check if a string can become empty by recursive deletion

#Input  : str = "GEEGEEKSKS", sub_str = "GEEKS"
#Output : Yes
#Explanation : In the string GEEGEEKSKS, we can first 
#              delete the substring GEEKS from position 4.
#              The new string now becomes GEEKS. We can 
#              again delete sub-string GEEKS from position 1. 
#              Now the string becomes empty.

def checkEmpty(input, pattern): 
      
    # If both are empty  
    if len(input)== 0 and len(pattern)== 0: 
         return 'true'
  
    # If only pattern is empty 
    if len(pattern)== 0: 
         return 'true'
  
    while (len(input) != 0): 
  
          # find sub-string in main string 
          index = input.find(pattern) 
            
          # check if sub-string founded or not 
          if (index ==(-1)): 
              return 'false'
            
          # slice input string in two parts and concatenate 
          input = input[0:index] + input[index + len(pattern):]              
    return 'true'
  
# Driver program 
if __name__ == "__main__": 
    input ='GEEGEEKSKS'
    pattern ='GEEKS'
    print(checkEmpty(input, pattern) )
    
##Python Counter| Find all duplicate characters in string
print('===Python Counter| Find all duplicate characters in string')

#Input : hello
#Output : l

from collections import Counter 
  
def find_dup_char(input): 
  
    # now create dictionary using counter method 
    # which will have strings as key and their  
    # frequencies as value 
    WC = Counter(input) 
    j = -1
      
      
    # Finding no. of  occurrence of a character 
    # and get the index of it. 
    for i in WC.values(): 
        j = j + 1
        if( i > 1 ): 
            print(WC.keys(),[j] )
  
# Driver program 
if __name__ == "__main__": 
    input = 'geeksforgeeks'
    find_dup_char(input)     
```

```python
print('\n=====================DICTIONARY PROGRAMS==================\n')


##Python Program for Binary Search (Recursive and Iterative)
##recursion is simply a method call in which the method being called is the same as the one 
##making the call while iteration is when a loop is repeatedly executed until a certain condition is met.

'''In a nutshell, this search algorithm takes advantage of a collection of elements that is already sorted by ignoring half of the elements after just one comparison.

Compare x with the middle element.
If x matches with the middle element, we return the mid index.
Else if x is greater than the mid element, then x can only lie in the right (greater) half subarray after the mid element. Then we apply the algorithm again for the right half.
Else if x is smaller, the target x must lie in the left (lower) half. So we apply the algorithm for the left half.'''

##Find 10 in the input array
arr = [ 2, 3, 4, 10, 40, 35 ] 
x = 40

print('input string is : ', arr)

##Recursive - 

def recursion_find (arr, l, r, x): 
  
    # Check base case 
    if r >= l: 
  
        mid = l + (r - l)//2
  
        # If element is present at the middle itself 
        if arr[mid] == x: 
            return mid 
          
        # If element is smaller than mid, then it can only 
        # be present in left subarray 
        elif arr[mid] > x: 
            return recursion_find(arr, l, mid-1, x) 
  
        # Else the element can only be present in right subarray 
        else: 
            return recursion_find(arr, mid+1, r, x) 
  
    else: 
        # Element is not present in the array 
        return -1
        

print('Below is recurive way')
a = recursion_find(arr,0,len(arr)-1,x)
if a != -1:
    print(x," is found in input string  and the position in sorted array is :", a)
else:
    print(x," is not found in input string")

##Iterative -     
def iterative_find(arr, l, r, x): 
  
    while l <= r: 
  
        mid = l + (r - l)//2; 
          
        # Check if x is present at mid 
        if arr[mid] == x: 
            return mid 
  
        # If x is greater, ignore left half 
        elif arr[mid] < x: 
            l = mid + 1
  
        # If x is smaller, ignore right half 
        else: 
            r = mid - 1
      
    # If we reach here, then the element was not present 
    return -1

print('Below is iterative way')
b = iterative_find(arr, 0, len(arr)-1, x) 

if b != -1:
    print(x," is found in input string  and the position in sorted array is :", b)
else:
    print(x," is not found in input string")
    
##Python Program for Insertion Sort
##Insertion sort is a simple sorting algorithm that works the way we sort playing cards in our hands.

def insertionSort(arr): 
  
    # Traverse through 1 to len(arr) 
    for i in range(1, len(arr)): 
  
        key = arr[i] 
  
        # Move elements of arr[0..i-1], that are 
        # greater than key, to one position ahead 
        # of their current position 
        j = i-1
        while j >=0 and key < arr[j] : 
                arr[j+1] = arr[j] 
                j -= 1
        arr[j+1] = key 
    
  
  
# Driver code to test above 
arr = [12, 11, 13, 5, 6] 

print ("Sorted array is:") 
for i in range(len(arr)): 
    print ("%d" %arr[i]) 
    
```

```python
def insertionSortRecursive(arr,n):
    # base case
    if n<=1:
        return
     
    # Sort first n-1 elements
    insertionSortRecursive(arr,n-1)
    ##\'\'\'Insert last element at its correct position
    ##    in sorted array.\'\'\'
    last = arr[n-1]
    print('last',last)
    j = n-2
    print('j',j)
    print(arr[j])
     
      # Move elements of arr[0..i-1], that are
      # greater than key, to one position ahead
      # of their current position 
    while (j>=0 and arr[j]>last):
        arr[j+1] = arr[j]
        j = j-1
 
    arr[j+1]=last
    
def printArray(arr,n):
    for i in range(n):
        print('printing',arr[i])
 
# Driver program to test insertion sort 
arr = [12,11,13,5,6]
n = len(arr)
print('n',n)
insertionSortRecursive(arr, n)
printArray(arr, n)

###Iterating over a string

lst = 'abcd'
  
iter_lst = iter(lst) 
while True: 
    try: 
        print(iter_lst.__next__())  
    except: 
        break
        
####Understand Pass by Value vs Pass by Refernce
## a variable is a refernce to some object in python

def swap(x, y): 
    temp = x; 
    x = y; 
    y = temp; 
    return x,y
  
# Driver code 
x = 2
y = 3
t = swap(x, y) 
print(x) 
print(y) 
print(type(t))
a,b = swap(x, y) 
print(a)
print(b)



# closures 
## A Closure is a function object that remembers values in enclosing scopes even if they are not present in memory.
##A closure—unlike a plain function—allows the function to access those captured variables through the closure’s 
####copies of their values or references, even when the function is invoked outside their scope.
def outerFunction(text): 
    text = text 
  
    def innerFunction(): 
        print(text) 
  
    return innerFunction # Note we are returning function WITHOUT parenthesis 
  
if __name__ == '__main__': 
    myFunction = outerFunction('Hey!') 
    myFunction() ## here closure is able to access the inner function
    print(myFunction)##you will get warning
    
    
##Factorial without Decorator

# Simple recursive program to find factorial 
def facto(num): 
    if num == 1: 
        return 1
    else: 
        return num * facto(num-1) 
          
  
print(facto(5)) 

###Factorial With Decorator

def memoize_factorial(f): 
    memory = {} 
  
    # This inner function has access to memory 
    # and 'f' 
    def inner(num): 
        if num not in memory:          
            memory[num] = f(num) 
        return memory[num] 
  
    return inner 
      
@memoize_factorial
def facto(num): 
    if num == 1: 
        return 1
    else: 
        return num * facto(num-1) 
  
print(facto(5)) 

##Help

help(facto)
```
